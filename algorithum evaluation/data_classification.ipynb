```python
x=df.drop(["id","loan_status"],axis=1)
y=df["loan_status"]
```

```python
cat=x.columns[x.dtypes=="object"]
num=x.columns[x.dtypes!="object"]
```

```python
#create pipeline
from sklearn.pipeline import Pipeline

from sklearn.preprocessing import StandardScaler,OneHotEncoder,OrdinalEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
sns.barplot(y.value_counts())
```

```python
cat_pipeline  =  Pipeline(steps  =  [('imputer' ,  SimpleImputer(strategy  =  'most_frequent')) ,  
                 ('encoder' ,  OrdinalEncoder())])

num_pipeline  =  Pipeline(steps  = [('imputer' ,  SimpleImputer(strategy  =  'mean' )) ,
                ('scaler' ,  StandardScaler())])
```

```python
#compose all pipelines
pre=ColumnTransformer([("cat",cat_pipeline,cat),
                   
                   ("num",num_pipeline,num)
                   

                   ])
```

```python
pre
```

```python
x_scaled=pre.fit_transform(x)
```

```python
x_scaled
```

```python
x_scaled=pd.DataFrame(x_scaled,columns  =  pre.get_feature_names_out())
```

```python
x_scaled
```

```python
### train test split
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=3)
```

```python
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
import xgboost
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.model_selection import cross_val_score
```

```python
lr= LogisticRegression()
lr.fit(x_train,y_train)
cvs_lr=cross_val_score(lr,x_train,y_train, cv=5)
dt= DecisionTreeClassifier(max_depth=6)
dt.fit(x_train,y_train)
cvs_dt=cross_val_score(lr,x_train,y_train, cv=5)
rf= RandomForestClassifier(max_depth=6)
rf.fit(x_train,y_train)
cvs_rf=cross_val_score(lr,x_train,y_train, cv=5)
gb=GradientBoostingClassifier(max_depth=6)
gb.fit(x_train,y_train)
cvs_gb=cross_val_score(lr,x_train,y_train, cv=5)
xgb=XGBClassifier(max_depth=6)
xgb.fit(x_train,y_train)
cvs_xgb=cross_val_score(lr,x_train,y_train, cv=5)
knn=KNeighborsClassifier(7)
knn.fit(x_train,y_train)
cvs_knn=cross_val_score(lr,x_train,y_train, cv=5)
svm=SVC()
svm.fit(x_train,y_train)
cvs_svm=cross_val_score(lr,x_train,y_train, cv=5)



```

```python
print("cvs_lr " , cvs_lr.mean())
print("cvs_dt " , cvs_dt.mean())
print("cvs_rf " , cvs_rf.mean())
print("cvs_gb " , cvs_gb.mean())
print("cvs_xgb",cvs_xgb.mean())
print("cvs_knn",cvs_knn.mean())
print("cvs_svm",cvs_svm.mean())
```

```python
def evaluate(model   ,  x , y):
    pred  =  model.predict(x)
    f1 =  f1_score(y ,  pred , average  = 'macro')
    cr  =  classification_report(y  , pred)
    cf  =  confusion_matrix(y ,  pred)
 
    cv  =  cross_val_score( model  , x ,  y ,  n_jobs= -1 , scoring= 'f1_macro' , cv  =  5 )
   
    print(cv.mean())
    print(f'f1-->{f1}')
    print(f'classification report -->{cr}')
    print(sns.heatmap(cf  ,  annot  = True  , fmt  = 'd'))
```

```python
from sklearn.metrics import (classification_report,confusion_matrix,f1_score)
```
