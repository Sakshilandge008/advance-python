```python
# create function

def div1():
    try:
        num=float(input("enter no here"))
        den=float(input("enter no here"))
        div=num/den
        print(f"division of {num} and {den} --->{num}/{den}")
        return div

    except Exception as e:
        print(f"exception occured at{e}")
        
```

```python
a=div1()
print(a)
```

```python
# use customize message for division handling

# create function

def div2():
    try:
        num=float(input("enter no here"))
        den=float(input("enter no here"))
        div=num/den
        print(f"division of {num} and {den} --->{num}/{den}")
        return div
    except ZeroDivisionError:
        print("denominator should not be 0")
    except ValueError:
        print("input should not be string or blank")
    except Exception as e:
        print("Exception occoured at {e}")        
```

```python
div2()
```

```python
def  file_read(path):
    try:
        import pandas as pd
        df =  pd.read_csv(path)
        return  df
    except ImportError:
        print("file could not import properly")
    except FileNotFoundError :
        print("file not found") 
    except ModuleNotFoundError:
        print("module not found")
    except Exception as e:
        print("exception occoured at {e}")          
```

```python
def square():
    try:
        num=float(input())
        s = num*num
    except Exception as e:
        print(e)
    else:
        print(f'square of {num} is {num}*{num}')
        return s
    finally:
        print("i will always execute")    
```

```python
square()
```
