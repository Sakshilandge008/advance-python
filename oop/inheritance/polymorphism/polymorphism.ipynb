```python
# create class rectangele , circle --> perimeter ,area
```

```python
class rectangle():
    def __init__(self, length, width):
        self.length = length
        self.width = width
    def perimeter(self):

        return 2*(self.length + self.width)
       
    def area(self):
        return self.length* self.width
class circle():
    def __init__(self,radius):
        self.radius = radius
        
    def perimeter(self):

         return 3.14 * (self.radius ** 2)
       
    def area(self):
        return 2 * 3.14 * self.radius
    
        
```

```python
p1=rectangle(2,3)
```

```python
p1.perimeter()
```

```python
p1.area()
```

```python
r1=circle(radius=2)
```

```python
r1.area()
```

```python
r1.perimeter()
```

```python
for i in [r1,p1]:
    print(i)
    print(f"perimeter-->{i.perimeter()}")
    print(f'area ---> {i.area()}')
```

```python
##encapsulation
# public variable---> access publically

#protected --> _ is used
#private ---> __ is used
```

```python
class acount:
    def __init__(self,balance):
        self.balance=balance
        
        # protected variable
        
```

```python
class acount1:
    def __init__(self,balance):
        self._balance=balance
        
```

```python
a1=acount1(20000)
```

```python
a1._balance
```

```python
#private variables

class acount2:
    def __init__(self,balance):
        self.__balance=balance
```

```python
a2=acount2(35433)
```

```python
a2.__balance
```

```python
a2._acount2__balance
```
