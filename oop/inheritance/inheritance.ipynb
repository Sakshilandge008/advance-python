```python
class person:
     def __init__(self, name, age):
          self.name=name
          self.age=age
     def introduce(self):
        return f"my name is {self.name} and i am {self.age} years old"      

# employee
class employee(person):
    def __init__(self, name, age,company):
        super().__init__(name, age)
        self.company=company

    def info(self):
        
        return f'{self.introduce()} i work for {self.company}'        
```

```python
p1=person(name="sakshi",age=20)
```

```python
p1.introduce()
```

```python
e1=employee(name="sakshi",age=20 , company="tcs")
```

```python
e1.info()
```

```python

#multilevel inheritance

## create classes employee , manager , project manager
class employee:
    def __init__(self,name,emp_id):
        self.name=name
        self.emp_id=emp_id
    def get_employee_info(self):
        return f'name:{self.name}, id: {self.emp_id}'

class manager(employee):
    def __init__(self, name, emp_id,department):
        super().__init__(name, emp_id)
        self.department=department

    def get_manager_info(self):
        return f"{self.get_employee_info()}, department:{self.department}"

class project_manager(manager):
    def __init__(self, name, emp_id, department,project):
        super().__init__(name, emp_id, department)
        self.project=project

    def get_project_manager_info(self):
        return f"{self.get_manager_info()},project : {self.project}"    
```

```python
e1=employee(name="sak",emp_id=2)
```

```python
e1.get_employee_info()
```

```python
m1=manager(name="sakshi",emp_id=12,department="ds")
```

```python
m1.get_manager_info()
```

```python
pr=project_manager(name="hruday",emp_id=4,department="ink",project="ink project")
```

```python
pr.get_project_manager_info()
```

```python
#hierarchical inheritance

## create class user as parent and regular user as child classes
```

```python
class user:
    def __init__(self,username,password):
        self.username=username
        self.password=password

    def get_user_info(self):
        return f"username={self.username} , password={self.password}"

class admin(user):
    def __init__(self, username, password,access):
        super().__init__(username, password)
        self.access=access

    def get_admin_info(self):
        return f"{self.get_user_info()}, access={self.access}"    
    
class regular_user(user):
    def __init__(self, username, password,subscription):
        super().__init__(username, password)
        self.subscription=subscription

    def get_regular_user_info(self):
        return f"{self.get_user_info()}, subsciption={self.subscription}"    


    
```

```python
u1=user("username=","sakshi")
```

```python
u1.get_user_info()
```

```python
a1=admin("sarthak","sarthak@12","complete")
```

```python
a1.get_admin_info()
```

```python
r1=regular_user("sarthak","sarthak@12",True)
```

```python
r1.get_regular_user_info()
```

```python
from pydantic import BaseModel,EmailStr , Field
from typing import Literal ,List
```

```python
#crete class student by base model
name:str
roll_no:int
marks: float
email: EmailStr
gender: Literal["male","female"]
hobbies:list[str]
```

```python
class student(BaseModel):
    name:str=Field(description ="enter name here")
    roll_no:int=Field(description="enter roll no ",gt=0)
    marks: float =Field(description="enter marks",gt=0,lt=100)
    email: EmailStr=Field(description = "enter name here", max_length=50)
    gender: Literal["male","female"]=Field(description="enter gender")
    hobbies:list[str]=Field(description="enter hobbies",default_factory=list)

    #create method
    def student_into(self):
        print(f"i am :{self.name}")
        print(f"roll no: {self.roll_no}")
        print(f"marks: {self.marks}")
        print(f"gender: {self.gender}")
        print(f"hobbies: {self.hobbies}")


```

```python
s1=student(name="hruday",roll_no=12, marks=60,email="example@gmail.com",gender="male", hobbies=["swimming"])
```

```python
s1.student_into()
```
