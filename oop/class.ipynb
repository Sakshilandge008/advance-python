```python

from dataclasses import dataclass
from typing import Literal ,List
```

```python
@dataclass
class student:
 name  : str
 roll_no  :  int
 marks  :  float
 gender  :  Literal['male' ,  'female' ]
 Hobbies  :  List[str]


 ##add student info method
 def info(self):
        print(f'i am {self.name}')   
        print(f'my roll no {self.roll_no} ' )
        print(f'my marks is {self.marks}')
        print(f'i am {self.gender}')
        print(f'i work as {self.Hobbies}')  
```

```python
## create class account with 
# check balance, deposite , withdraw, transfer methods
# numberm name, balance --attr
```

```python
@dataclass
class account:
    acc_no:int
    acc_name:str
    balance:float
#create ,method check balance
    def check_balance(self):
     return f'current balance of account {self.acc_no} is {self.acc_name} and balance is {self.balance}'   


    # add money to acc deposite
    def deposite(self,amt:float):
       self.balance+=amt
       print(f'amount of {amt} is deposited in {self.acc_no}')
       return self.check_balance()

    #withdraw amount
    def withdraw(self,amt):
       if self.balance>=amt:
          self.balance-=amt
          print(f'amount if {amt} is withdrawn from {self.acc_no}')
          return self.check_balance() 
       else:
          raise ValueError("insufficient funds")
       
    # transfer money
    def transfer(self,amt:float,acc='account'):
      self.withdraw(amt)
      acc.deposite(amt)   
      return self.check_balance(),acc.check_balance()

```

```python
a1=account(acc_no=22,acc_name="xyzt",balance=200050)
```

```python
a2=account(acc_no=21,acc_name="xyz",balance=20000)
```

```python
a1.transfer(amt=20000,acc=a2)
```

```python
#create class to find area and perimeter of triangle

from dataclasses import dataclass
```

```python
@dataclass
class triangle:
    s1:int|float
    s2:int|float
    s3:int|float

    def perimeter(self):
        if (self.s1+self.s2>self.s3) and\
        (self.s2+self.s3>self.s1) and\
        (self.s1+self.s3>self.s2):
          p=self.s1+self.s2+self.s3
          return p
        else:
           print("triangle not formed")

    #find area of triangle
    # 
    def area(self):
       if self.perimeter() is not None:
          
        s=self.perimeter()
        return(s*((s-self.s1)*(s-self.s2)*(s-self.s3)))**(1/2)    
       
            
```

```python
triangle(s1=2,s2=3,s3=3)
```

```python
t1=triangle(2,3,3)
```

```python
t1.perimeter()
```
