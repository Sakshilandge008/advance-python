```python
def info(**kwargs):
    user={"name":None,
          "age":None,
          "email":None}
    for key in kwargs.keys():
        if key in user.keys():
            user[key]=kwargs[key]
    return user        
```

```python
#  create decorator  to welcome  the  user
def  Welcome(func):
  def wrapper(*args  , **kwargs):
    print('Welcome user')
    result  =  func(*args  , **kwargs)
    print(f'Result  of  function is {result}')
    print('Thank  you')

    return result
  return  wrapper
```

```python
#create function for factorial of no apply decorator

@welcome
def factorial(num):
    fact=1
    for i in range(1,num+1):
        fact=fact +1
    return factorial    

```
